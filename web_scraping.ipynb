```python
import pandas as pd
import matplotlib.pyplot as plt
from pytrends.request import TrendReq
```

```python
trends = TrendReq()
trends
```

```python
kw_list = ["Machine Learning"]
trends.build_payload(kw_list, cat=0, timeframe='today 5-y', geo='', gprop='')
data = trends.interest_by_region()
data = data.sort_values(by="Machine Learning", ascending=False)
data = data.head(10)
```

```python
data
```

# Visualization

```python
data.reset_index().plot(x="geoName", y="Machine Learning",figsize=(15,5), kind='bar')
plt.show()
```

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=["Machine Learning"])
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(18,15))
data["Machine Learning"].plot()
plt.style.use("fivethirtyeight")
plt.title("Google search result for Machine Learning for  4 years")
plt.show()
```
